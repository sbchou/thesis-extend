```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas
%matplotlib inline
import numpy as np 
```

```python
from newspaper import Article
import csv
from collections import Counter
import re
```

```python
with open('DATA/EMOT/Negativ.csv', 'rb') as f:
    reader = csv.reader(f, quotechar='"')
    negativ = list(reader)
```

```python
with open('DATA/EMOT/Positiv.csv', 'rb') as f:
    reader = csv.reader(f, quotechar='"')
    positiv = list(reader)
```

```python
positiv = [p[0] for p in positiv]
negativ = [n[0] for n in negativ]
```

```python
negativ
```

```python
def get_story(url):
    article = Article(url)
    article.download()
    article.parse()
    body = article.text
    return body.upper()
    
```

```python
def count_negativ(body): 
    count = Counter(re.findall("[A-Z]{2,}(?![a-z])|[A-Z][a-z]+(?=[A-Z])|[\'\w\-]+", body))
    return len({key: count[key] for key in count if key in negativ}.values())
```

```python
def count_positiv(body): 
    count = Counter(re.findall("[A-Z]{2,}(?![a-z])|[A-Z][a-z]+(?=[A-Z])|[\'\w\-]+", body))
    return len({key: count[key] for key in count if key in positiv}.values())
```

```python
def wc(body): 
    return len(re.findall("[A-Z]{2,}(?![a-z])|[A-Z][a-z]+(?=[A-Z])|[\'\w\-]+", body))
```

```python
clinton_bott = pandas.read_csv('DATA/STANCE_VIRALITY/Clinton_bottom10url.csv', names=['url', 'num_tweets'])
```

```python
clinton_bott['body'] = clinton_bott.url.apply(lambda url: get_story(url))
```

```python
clinton_bott['wc'] = clinton_bott.body.apply(lambda body: wc(body))
```

```python
clinton_bott['negativ_ct'] = clinton_bott.body.apply(lambda body: count_negativ(body))
```

```python
clinton_bott['positiv_ct'] = clinton_bott.body.apply(lambda body: count_positiv(body))
```

```python
clinton_bott
```

```python
 clinton_top = pandas.read_csv('DATA/STANCE_VIRALITY/Clinton_top10url.csv', names=['url', 'num_tweets'])
```

```python
clinton_top['body'] = clinton_top.url.apply(lambda url: get_story(url))
```

```python
clinton_top['wc'] = clinton_top.body.apply(lambda body: wc(body))
```

```python
clinton_top['negativ_ct'] = clinton_top.body.apply(lambda body: count_negativ(body))
```

```python
clinton_top['positiv_ct'] = clinton_top.body.apply(lambda body: count_positiv(body))
```

```python
clinton_top['positiv_percent'] = clinton_top.positiv_ct / (1.0 * clinton_top.wc)
```

```python
clinton_top['negativ_percent'] = clinton_top.negativ_ct / (1.0 * clinton_top.wc)
```

```python
clinton_bott['positiv_percent'] = clinton_bott.positiv_ct / (1.0 * clinton_bott.wc)
```

```python
clinton_bott['negativ_percent'] = clinton_bott.negativ_ct / (1.0 * clinton_bott.wc)
```

```python
clinton_bott['emotionality'] = (clinton_bott.negativ_ct + clinton_bott.positiv_ct) / (1.0 * clinton_bott.wc)
```

```python
clinton_top['emotionality'] = (clinton_top.negativ_ct + clinton_top.positiv_ct) / (1.0 * clinton_top.wc)
```

```python
clinton_bott['positivity'] = clinton_bott.positiv_percent - clinton_bott.negativ_percent
```

```python
clinton_top['positivity'] = clinton_top.positiv_percent - clinton_top.negativ_percent
```
